```python
import pandas as pd
import numpy as np
import codecs
from scipy import linalg

import make_database as md
import make_bit_frame as mbf
import make_distance_matrix as mdm
import generate_controler as gc
from AIC import *

import glob
import copy
from __future__ import division
```

```python
fileList = glob.glob('./csvFileList/*.csv')
gcc = gc.GenerateControler(fileList, "./InputValueDB/母音入力数値一覧.csv", "./InputValueDB/子音入力数値一覧.csv")
```

```python
cons_col = (gcc.cons.columns == "子音").tolist().index(True)
# 最初の調音アライメント列 (母音も同じ)
first_columns = (gcc.cons.columns == "位置").tolist().index(True)

#locate_index = (self.table.T.iloc[cons_col, :] == ).tolist().index(True)
try:
    nan_position = (gcc.cons.T.iloc[cons_col, :] == gcc.cons.T.iloc[cons_col, :]).tolist().index(False)
    index_size = nan_position
except Exception as e:
    index_size = gcc.cons.index.size
    
art_lists = []
cons_name_lists = []
for kk in range(0, index_size):
    art_lists.append(gcc.cons.loc[kk][first_columns:first_columns + 5].tolist())
    cons_name_lists.append(gcc.cons.loc[kk][cons_col])
```

```python
vowel_col = (gcc.vowel.columns == "母音").tolist().index(True)
# 最初の調音アライメント列 (母音も同じ)
first_columns = (gcc.vowel.columns == "長さ").tolist().index(True)

#locate_index = (self.table.T.iloc[cons_col, :] == ).tolist().index(True)
try:
    nan_position = (gcc.vowel.T.iloc[vowel_col, :] == gcc.vowel.T.iloc[vowel_col, :]).tolist().index(False)
    index_size = nan_position
except Exception as e:
    index_size = gcc.vowel.index.size
    
art_lists = []
vowel_name_lists = []
for kk in range(0, index_size):
    art_lists.append(gcc.vowel.loc[kk][first_columns:first_columns + 5].tolist())
    vowel_name_lists.append(gcc.vowel.loc[kk][vowel_col])
```

```python
count_cons_changed = pd.DataFrame([[0 for i in range(len(cons_name_lists))] for k in range(len(cons_name_lists))], cons_name_lists, cons_name_lists)
count_vowel_changed = pd.DataFrame([[0 for i in range(len(vowel_name_lists))] for k in range(len(vowel_name_lists))], vowel_name_lists, vowel_name_lists)
```

```python
def makeFrameList(table):
    # region_col = (table.loc[0, :] == "表示用").tolist().index(True)  # 表示用のcolumns番号
    locate_col = (table.loc[0, :] == "現市町村名").tolist().index(True)  # 表示用のcolumns番号    
    number_col = (table.loc[0][:] == "通し番号").tolist().index(True)
    try:
        nan_position = (table.T.iloc[number_col, :] == table.T.iloc[number_col, :]).tolist().index(False)
        index_size = nan_position
    except Exception as e:
        index_size = table.index.size

    region_col = (table.loc[0, :] == "表示用").tolist().index(True)
    try:
        first_of_index = (table.T.iloc[number_col, :] == "000").tolist().index(True)  # 000の最初のindex番号
    except Exception as e:
        first_of_index = (table.T.iloc[number_col, :] == "0").tolist().index(True)

    locate_index = (table.T.iloc[locate_col, :] == "想定形").tolist().index(True)  # 001の最初のindex番号
    syllable_col = (table.loc[0, :] == "モーラ").tolist().index(True)  # 音節のcolumns番号
    max_syllable_num = max([int(x) if (x == x) == True else 0 for x in table.T.iloc[syllable_col, 1:index_size].tolist()])

    first_cons = (table.loc[0, :] == "1子音").tolist().index(True)  # 最初の子音列
    # 最初の調音アライメント列 (母音も同じ)
    first_columns = (gcc.cons.columns == "位置").tolist().index(True)
    souteikei_list = [0]*2*max_syllable_num
    for kk in range(locate_index, index_size):
        if(table.loc[kk][region_col] == "岩波" or table.loc[kk][region_col] == "鹿児島" or table.loc[kk][region_col] == "種子" or table.loc[kk][region_col] == "八丈" or table.iloc[kk][region_col] == "OCLS"):
            continue                    
        for k in range(first_cons, first_cons + 2 * max_syllable_num):            
            if (k % first_cons) % 2 == 0:
                try:
                    index = (table.T.iloc[k][kk] == gcc.cons.iloc[:]["子音"]).tolist().index(True)
                except Exception as e:
                    print("第{0}列目の子音:{1}がデータベース内から見つかりませんでした。".format(kk,table.T.iloc[k][kk]))
                    return(-1)
                    # sys.exit()
                pron = table.T.iloc[k][kk]
                if kk == locate_index:                    
                    souteikei_list[k-first_cons] = pron
                else:
                    count_cons_changed.loc[souteikei_list[k-first_cons]][pron] +=1
                    count_cons_changed.loc[pron][souteikei_list[k-first_cons]] +=1
            else:
                try:
                    index = (table.T.iloc[k][kk] == gcc.vowel.iloc[:]["母音"]).tolist().index(True)
                except Exception as e:
                    print("第{0}列目の, 母音:{1}がデータベース内から見つかりませんでした。".format(kk,table.T.iloc[k][kk]))
                    return(-1)
                    # sys.exit()
                pron = table.T.iloc[k][kk]
                if kk == locate_index:
                    souteikei_list[k-first_cons] = pron
                else:
                    count_vowel_changed.loc[souteikei_list[k-first_cons]][pron] +=1
                    count_vowel_changed.loc[pron][souteikei_list[k-first_cons]] +=1                               
    return 0
```

```python
count_cons_changed = pd.DataFrame([[0 for i in range(len(cons_name_lists))] for k in range(len(cons_name_lists))], cons_name_lists, cons_name_lists)
count_vowel_changed = pd.DataFrame([[0 for i in range(len(vowel_name_lists))] for k in range(len(vowel_name_lists))], vowel_name_lists, vowel_name_lists)
for fl in range(0, len(fileList)):
    with codecs.open(fileList[fl], "r", "Shift-JIS", "ignore") as files:
        mtoshi = pd.read_table(files, delimiter=",")
    print(fileList[fl])
    makeFrameList(mtoshi)
```
