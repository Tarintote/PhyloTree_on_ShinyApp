```python
import pandas as pd
import numpy as np
import codecs
from scipy import linalg

import make_database as md
import make_bit_frame as mbf
import make_distance_matrix as mdm
import generate_controler as gc
from AIC import *

import glob
import copy
from __future__ import division
```

```python
fileList = glob.glob('./csvFileList/*.csv')
gcc = gc.GenerateControler(fileList, "./InputValueDB/母音入力数値一覧.csv", "./InputValueDB/子音入力数値一覧.csv")
```

```python
gcc.mainPreProccess()
```

```python
cons_col = (gcc.cons.columns == "子音").tolist().index(True)
# 最初の調音アライメント列 (母音も同じ)
first_columns = (gcc.cons.columns == "位置").tolist().index(True)

#locate_index = (self.table.T.iloc[cons_col, :] == ).tolist().index(True)
try:
    nan_position = (gcc.cons.T.iloc[cons_col, :] == gcc.cons.T.iloc[cons_col, :]).tolist().index(False)
    index_size = nan_position
except Exception as e:
    index_size = gcc.cons.index.size
    
art_lists = []
name_lists = []
for kk in range(0, index_size):
    art_lists.append(gcc.cons.loc[kk][first_columns:first_columns + 5].tolist())
    name_lists.append(gcc.cons.loc[kk][cons_col])

mDM = mdm.MakeDistanceMatrix()
mDM.makeHammDistanceFrame(art_lists)
cons_data_frame = mDM.getDistanceFrame()
```

```python
pd.DataFrame(cons_data_frame, name_lists, name_lists).to_csv('./SpecificCSVFiles/distance/cons_dis.csv', sep=',')
```

```python
vowel_col = (gcc.vowel.columns == "母音").tolist().index(True)
# 最初の調音アライメント列 (母音も同じ)
first_columns = (gcc.vowel.columns == "長さ").tolist().index(True)

#locate_index = (self.table.T.iloc[cons_col, :] == ).tolist().index(True)
try:
    nan_position = (gcc.vowel.T.iloc[vowel_col, :] == gcc.vowel.T.iloc[vowel_col, :]).tolist().index(False)
    index_size = nan_position
except Exception as e:
    index_size = gcc.vowel.index.size
    
art_lists = []
name_lists = []
for kk in range(0, index_size):
    art_lists.append(gcc.vowel.loc[kk][first_columns:first_columns + 5].tolist())
    name_lists.append(gcc.vowel.loc[kk][vowel_col])

mDM = mdm.MakeDistanceMatrix()
mDM.makeHammDistanceFrame(art_lists)
vowel_data_frame = mDM.getDistanceFrame()
```

```python
pd.DataFrame(vowel_data_frame, name_lists, name_lists).to_csv('./SpecificCSVFiles/distance/vowel_dis.csv', sep=',')
```
